```python
# making the last few outputs into a for loop to get the differential

list_random = [random.randint(0,30) for x in range(0,100)]
list_divisible = []

for x in list_random:
    if  x % 3 == 0:
        list_divisible.append(x)
    list_random_len = len(list_random)
    list_divisible_len = len(list_divisible)
    list_diff1 = list_random_len - list_divisible_len
print(list_diff1)
```

```python
# repeat process for 2nd time

list_random = [random.randint(0,30) for x in range(0,100)]
list_divisible = []

for x in list_random:
    if  x % 3 == 0:
        list_divisible.append(x)
    list_random_len = len(list_random)
    list_divisible_len = len(list_divisible)
    list_diff2 = list_random_len - list_divisible_len
print(list_diff2)
```

```python
# repeat process for 3rd time

list_random = [random.randint(0,30) for x in range(0,100)]
list_divisible = []

for x in list_random:
    if  x % 3 == 0:
        list_divisible.append(x)
    list_random_len = len(list_random)
    list_divisible_len = len(list_divisible)
    list_diff3 = list_random_len - list_divisible_len
print(list_diff3)
```

```python
# creating a blank string to which I attach all the words without special characters

no_spec_char_multiline_text = ''
for char in multiline_text:
    if char == ' ':
        no_spec_char_multiline_text += char
    elif char.isalnum():
        no_spec_char_multiline_text += char
    else:
        no_spec_char_multiline_text += " "

        
# checking my work
print(no_spec_char_multiline_text)
```

```python
# creating a list of all of the word by using the split function

all_words = no_spec_char_multiline_text.split()
all_words
```

```python
# how many words

len(all_words)
```

```python
# making it a dictionary in order to track how many words are repeated

words_as_dict = dict.fromkeys(all_words)
```

```python
# list of words excluding the repeated words

len(list(words_as_dict.keys()))
```

```python
# creating a for loop that creates a word dictionary and counts how many times each is used

for word in all_words:
    if words_as_dict[word] is None:
        words_as_dict[word] = 1
    else:
        words_as_dict[word] += 1
words_as_dict
```

```python
# sort out the top 25 most used words of the dictionary

sorted_words = sorted(words_as_dict.items(), key=lambda key_val_tuple: key_val_tuple[1], reverse=True)
sorted_words[:25]
```

```python
# creating a combo_to_num function that takes the combinations of 0, 1, and 2 and makes a number

def combo_to_num(num_stack):
    last_num = 0
    for i in range(0,len(num_stack)):
        last_num += (num_stack.pop() * (math.pow(10, i)))
    return last_num

for combination_123 in permutations(range(3)):
    num_stack = list(dropwhile(lambda x: x<=0, combination_123))
    print(combo_to_num(num_stack))
```

```python
# creating a function that will take the csv and create a dictionary from it

def dict_from_csv(header, line):
    zipped_line = zip_longest(header, line, fillvalue = None)
    new_dict = {keyvalue[0]: keyvalue[1] for keyvalue in zipped_line}
    return new_dict
```
